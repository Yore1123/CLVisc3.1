# Glueball EOS

This note is used to test the Glueball EOS, from the plot we see that the upper limits of energy density and temperature are really high. There is first order phase transition at temperature 270 MeV. The energy density
around 1st order phase transition is empty in the table so we need to set constant energy density at T=270 MeV.
For temperature smaller than 189 MeV, we also need to do extrapolation.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

glueball_dat = np.loadtxt('glueball_eos.dat')

# T in units [GeV], others are dimensionless
T = 1.0E-3*glueball_dat[:,2]
e_o_T4 = glueball_dat[:,0] 
p_o_T4 = glueball_dat[:,4]
s_o_T3 = glueball_dat[:,3]
hbarc = 0.19732 

energy_density = e_o_T4 * T**4 /(hbarc**3)       # in units GeV/fm^3
pressure = p_o_T4 * T**4 / (hbarc**3)            # in units GeV/fm^3
entropy_density = s_o_T3 * T**3 / (hbarc**3)     # in units fm^{-3}
```

```python
plt.plot(energy_density, pressure, 'ro')
plt.xlabel('energy density [GeV/fm^3]')
plt.ylabel('pressure [GeV/fm^3]')
```

```python
# plot p(e) around first order phase transition
plt.plot(energy_density, pressure, 'ro')
plt.xlabel('energy density [GeV/fm^3]')
plt.ylabel('pressure [GeV/fm^3]')
plt.xlim(0, 5.0)
plt.ylim(0, 0.7)
```

```python
plt.plot(T, energy_density, 'ro')
plt.xlabel('temperature [GeV]')
plt.ylabel('energy density [GeV/fm^3]')
plt.xlim(0, 0.3)
plt.ylim(0, 5)
```
